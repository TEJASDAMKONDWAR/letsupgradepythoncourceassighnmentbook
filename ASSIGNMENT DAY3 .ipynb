```python
avg=int(input("write your marks here:"))
if(avg>=90):
    print("Grade: A")
elif(avg>=80&avg<90):
    print("Grade: B")
elif(avg>=70&avg<80):
    print("Grade: C")
elif(avg>=60&avg<70):
    print("Grade: D")
else:
    print("Grade: F")
```

# ASSIGNMENT 2

```python
A = 1 
B= 1000

print("Prime numbers between", A, "and", B, "are:")

for num in range(A, B + 1):
   
   if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            print(num)
```
